```python
#3.1
from sklearn.feature_extraction import text
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
my_stop_words = text.ENGLISH_STOP_WORDS

data = pd.DataFrame(train[['id','Name','Description']])
unique_data=data.drop_duplicates(['id'])[['id','Name','Description']]
#me to parapanw tropo petaw ta duplicate ids kai krataw mono ena discription gia ka8e ena wste na exw kalutera 
#apotelesmata, ginetai kai me unique alla einai pio periploko kai me groupby 
#kanw ta fillna an kai de nomizw oti xreiazontai apo kapoious elegxouw pou ekana
unique_data['Name'].fillna('NULL')
unique_data['Description'].fillna('NULL')
#ta enwnw se mia kainouria stili
unique_data['metadata'] = unique_data.Name + ' ' + unique_data.Description
#kanw drop ta palia columns, an mou xreiastoun isws valw se sxolio to parakatw kwdika
#unique_data.drop(columns = ['Name','Description'])

v = TfidfVectorizer(stop_words=my_stop_words)
arr = v.fit_transform(unique_data['metadata'][0:500].values.astype(str))        
arr = arr.toarray()    
arr
```

```python
#3.2
#8a parei 2-3 lepta
#dinw 1000 times giati apo ena simeio kai meta apla pairnei polu wra i diadikasia kai stamataei to kernel
from operator import itemgetter
#xrisimopoieitai gia na sortaroume listes apo dictionaries sunithws
csim = cosine_similarity(arr)
dictionary = [] 
xcounter = 0 #krataw ton ari8mo tis seiras pou vriskomaste sto pinaka
for x in csim:
    ycounter = 0 #krataw ton ari8mo tis stilis pou vriskomaste sto pinaka
    for y in x:
        #8a apo8ikeusw ta id pou sugkrinoume kai ti cosine_similarity tous se ena temporary dictionary
        #kai 8a ta apo8ikeusw ola sto dictionary gia na kanw meta tis apaitoumenes sugkriseis
        temp = {} 
        temp['id1'] = unique_data.id[xcounter] #xrisimopoiw to dataframe apo to proigoumeno paradeigma gia na 
        #antistoixisw katallila ta id sto dictionary
        temp['id2'] = unique_data.id[ycounter]
        temp['csim'] = csim[xcounter][ycounter]
        temp['Name1'] = unique_data.Name[xcounter]
        temp['Name2'] = unique_data.Name[ycounter]
        temp['Description'] = unique_data.Description[ycounter]
        dictionary.append(temp)
        ycounter += 1 
    xcounter += 1
    
dictionary = sorted(dictionary,key = itemgetter('csim'))
dictionary[0:100]
```

```python
#3.3
def recommend(item_id,num):
    counter = 0
    id1 = 0
    for item in dictionary: #xrisimopoiw to dictionary pou eftia3a sto proigoumeno erwtima kai 8a e3agw apo ekei
        #ta aparaitita dedomena
        if (item['id1'] == item_id and id1 == 0):
            print("Recommending 5 listings similar to",item['Name1'])
            print("---------------------------------------------------------")
            id1 = 1
        if (item['id1'] == item_id and counter < num):
            print("Recommended:",item['Name2'])
            print("Description:",item['Description'])
            print("(score:",item['csim'],")\n")
            counter += 1
        if counter > num:
            break
    if id1 == 0:
        print('This id doesnt exist in my base')
            
            
recommend(31155,4)
#ta score apo8ikeuontai kanonika alla exei 8ema sto na ta deixnei oloklira gia kapoio logo kai de vrika pws mporw na
#to alla3w auto, episis exei ena error to opoio de katalava ti akrivws pws dior8wnotan kai to periergo itan oti 
#trexei kanonika o kwdikas
```
